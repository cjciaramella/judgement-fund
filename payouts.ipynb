```python
# import libraries
import pandas as pd
import re
import numpy as np
```

```python
# import csv files and pass to variables
ohsix = pd.read_csv("2006-payments.csv", index_col = False, encoding = "ISO-8859-1")
ohseven = pd.read_csv("2007-payments.csv", index_col = False, encoding = "ISO-8859-1")
oheight = pd.read_csv("2008-payments.csv", index_col = False, encoding = "ISO-8859-1")
ohnine = pd.read_csv("2009-payments.csv", index_col = False, encoding = "ISO-8859-1")
ten = pd.read_csv("2010-payments.csv", index_col = False, encoding = "ISO-8859-1")
eleven = pd.read_csv("2011-payments.csv", index_col = False, encoding = "ISO-8859-1")
twelve = pd.read_csv("2012-payments.csv", index_col = False, encoding = "ISO-8859-1")
thirteen = pd.read_csv("2013-payments.csv", index_col = False, encoding = "ISO-8859-1")
fourteen = pd.read_csv("2014-payments.csv", index_col = False, encoding = "ISO-8859-1")
fifteen = pd.read_csv("2015-payments.csv", index_col = False, encoding = "ISO-8859-1")
sixteen = pd.read_csv("2016-payments.csv", index_col = False, encoding = "ISO-8859-1")
```

```python
frames = [ohsix, ohseven, oheight, ohnine, ten, eleven, twelve, thirteen, fourteen, fifteen, sixteen]
# get total rows
total = 0
for frame in frames:
    total += len(frame)
print(total)
```

```python
# concatenate DataFrames and get info
joined = pd.concat(frames)
joined.info()
```

```python
# export concatenated DataFrame to csv
joined.to_csv('joined-settlements.csv')
```

```python
# sort by agencies with most settlements
agency_count = joined["Defendant Agency Name"].value_counts().reset_index()
agency_count.rename(columns={'index': 'Agency', 'Defendant Agency Name': 'Settlements'}, inplace=True)
agency_count
```

```python
# export csv of agency_count
agency_count.to_csv('agency-count.csv')
```
